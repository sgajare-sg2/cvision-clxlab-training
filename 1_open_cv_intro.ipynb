## Making simple geometric shapes 

```python
import numpy as np
#We declared an empty canvas
canvas = np.zeros((300, 300, 3), dtype = "uint8")
```

```python
red = (255, 0, 0)
```

```python
cv2.line(canvas, (0,0), (300, 300), red)
```

```python
plt.imshow(canvas)
```

```python
green=(0, 255, 0)
blue=(0, 0, 255)
```

```python
cv2.rectangle(canvas, (10, 10), (60, 60), green, 5)
```

```python
plt.imshow(canvas)
```

```python
cv2.rectangle(canvas, (100, 100), (150, 250), blue, -1)
```

```python
plt.imshow(canvas)
```

Specifying the thickness as -1 filled the rectangle.

```python
canvas = np.zeros((300, 300, 3), dtype = "uint8")
#Made the canvas white
canvas[:,:] = (255, 255, 255)
#declared centre
centre=(canvas.shape[1]//2, canvas.shape[0]//2 )
#plotted circle in a loop
for r in range(0, 150, 10):
    cv2.circle(canvas, centre, r, blue)
plt.imshow(canvas)
```

```python
#Add this line incase you want the output to be identical at every run
#np.random.seed(42)
#Declaring and making a white canvas
canvas = np.zeros((300, 300, 3), dtype = "uint8")
canvas[:,:] = (255, 255, 255)
for i in range(0, 50):
    #Randomly create radii, color and center for a 50 circles
    radius= np.random.randint(5, high=100)
    color= np.random.randint(0, high=256, size=(3,)).tolist()
    centre = np.random.randint(0, high = 300, size = (2,))
    cv2.circle(canvas, tuple(centre), radius, color, -1)
plt.imshow(canvas)
```

```python
#Adding Text in the image
org=(150, 250)
font=cv2.FONT_HERSHEY_SIMPLEX
fontScale=1
color=(255, 255, 255)
thickness=2
cv2.putText(canvas, 'Cloudxlab', org, font, fontScale, color, thickness) 
plt.imshow(canvas)
```

### Creating picture with noise

We will show how to create an image with a single square and add some random noise

```python
canvas = np.zeros((300, 300), dtype = "uint8")
#Create a white square in the middle
canvas[100:200, 100:200]=255
#Randomly choose (x,y) and set pixel value to a random number in range 0,255
for i in range(9000):
    x=np.random.randint(0, high=300)
    y=np.random.randint(0, high=300)
    value=np.random.randint(0, high=256)
    canvas[x][y]=value
```

```python
canvas=cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB)
plt.imshow(canvas)
cv2.imwrite("images/noise.png", canvas)
```
